```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

%matplotlib inline
```

```python
# Load datasets

nav = pd.read_csv("../data/processed/nav_history_cleaned.csv")
fund = pd.read_csv("../data/01_fund_master.csv")
perf = pd.read_csv("../data/processed/scheme_performance_cleaned.csv")
txn = pd.read_csv("../data/processed/investor_transactions_cleaned.csv")
aum = pd.read_csv("../data/03_aum_by_fund_house.csv")
sip = pd.read_csv("../data/04_monthly_sip_inflows.csv")
category = pd.read_csv("../data/05_category_inflows.csv")
folio = pd.read_csv("../data/06_industry_folio_count.csv")
holding = pd.read_csv("../data/09_portfolio_holdings.csv")
benchmark = pd.read_csv("../data/10_benchmark_indices.csv")

print("All datasets loaded successfully!")
```

```python
# Display the first 5 rows of each dataset

print("Fund Master")
display(fund.head())

print("NAV History")
display(nav.head())

print("Scheme Performance")
display(perf.head())

print("Investor Transactions")
display(txn.head())

print("AUM")
display(aum.head())

print("Monthly SIP")
display(sip.head())

print("Category Inflows")
display(category.head())

print("Industry Folio")
display(folio.head())

print("Portfolio Holdings")
display(holding.head())

print("Benchmark Indices")
display(benchmark.head())
```

```python
# Check the dimensions of every dataset

datasets = {
    "Fund Master": fund,
    "NAV History": nav,
    "Scheme Performance": perf,
    "Investor Transactions": txn,
    "AUM": aum,
    "Monthly SIP": sip,
    "Category Inflows": category,
    "Industry Folio": folio,
    "Portfolio Holdings": holding,
    "Benchmark": benchmark
}

for name, df in datasets.items():
    print(f"{name}: {df.shape}")
```

```python
# Convert date column to datetime

nav["date"] = pd.to_datetime(nav["date"])

print(nav.head())
```

```python
# Pick first 5 mutual funds

top5 = nav["amfi_code"].unique()[:5]

print(top5)
```

```python
plt.figure(figsize=(14,6))

for code in top5:
    temp = nav[nav["amfi_code"] == code]
    plt.plot(temp["date"], temp["nav"], label=f"Fund {code}")

plt.title("Daily NAV Trend (Top 5 Funds)")
plt.xlabel("Date")
plt.ylabel("NAV")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# AUM Growth by Fund House

plt.figure(figsize=(12,6))

sns.barplot(
    data=aum,
    x="year",
    y="aum_crore",
    hue="fund_house"
)

plt.title("AUM Growth by Fund House")
plt.xlabel("Year")
plt.ylabel("AUM (Crore)")
plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("../reports/aum_growth.png", dpi=300)

plt.show()

print("AUM chart saved.")
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

os.makedirs("../reports", exist_ok=True)

# -------------------------------
# Chart 3 - SIP Inflow Trend
# -------------------------------
sip["month"] = pd.to_datetime(sip["month"])

plt.figure(figsize=(12,6))
plt.plot(sip["month"], sip["sip_inflow_crore"], marker='o', linewidth=2)
plt.title("Monthly SIP Inflows")
plt.xlabel("Month")
plt.ylabel("SIP Inflow (Crore)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../reports/sip_inflows.png", dpi=300)
plt.show()

# -------------------------------
# Chart 4 - Category Inflow Trend
# -------------------------------
category["month"] = pd.to_datetime(category["month"])

pivot = category.pivot(
    index="month",
    columns="category",
    values="net_inflow_crore"
)

plt.figure(figsize=(14,6))
sns.heatmap(pivot.T, cmap="YlGnBu")
plt.title("Category-wise Net Inflows")
plt.tight_layout()
plt.savefig("../reports/category_heatmap.png", dpi=300)
plt.show()

# -------------------------------
# Chart 5 - Gender Distribution
# -------------------------------
plt.figure(figsize=(6,6))
txn["gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.ylabel("")
plt.title("Investor Gender Distribution")
plt.tight_layout()
plt.savefig("../reports/gender_distribution.png", dpi=300)
plt.show()

# -------------------------------
# Chart 6 - Age Group Distribution
# -------------------------------
plt.figure(figsize=(8,5))
sns.countplot(data=txn, x="age_group")
plt.title("Investor Age Groups")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("../reports/age_distribution.png", dpi=300)
plt.show()

# -------------------------------
# Chart 7 - City Tier Distribution
# -------------------------------
plt.figure(figsize=(7,5))
sns.countplot(data=txn, x="city_tier")
plt.title("City Tier Distribution")
plt.tight_layout()
plt.savefig("../reports/city_tier_distribution.png", dpi=300)
plt.show()

# -------------------------------
# Chart 8 - Transaction Type
# -------------------------------
plt.figure(figsize=(6,5))
sns.countplot(data=txn, x="transaction_type")
plt.title("Transaction Types")
plt.tight_layout()
plt.savefig("../reports/transaction_types.png", dpi=300)
plt.show()

# -------------------------------
# Chart 9 - Folio Growth
# -------------------------------
folio["month"] = pd.to_datetime(folio["month"])

plt.figure(figsize=(12,6))
plt.plot(
    folio["month"],
    folio["total_folios_crore"],
    linewidth=2
)
plt.title("Industry Folio Growth")
plt.xlabel("Month")
plt.ylabel("Folios (Crore)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../reports/folio_growth.png", dpi=300)
plt.show()

# -------------------------------
# Chart 10 - Sector Allocation
# -------------------------------
sector = (
    holding.groupby("sector")["weight_pct"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,8))
plt.pie(
    sector,
    labels=sector.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Portfolio Sector Allocation")
plt.tight_layout()
plt.savefig("../reports/sector_allocation.png", dpi=300)
plt.show()

# -------------------------------
# Chart 11 - Top 10 Holdings
# -------------------------------
top_holdings = (
    holding.groupby("stock_name")["weight_pct"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))
sns.barplot(
    x=top_holdings.values,
    y=top_holdings.index
)
plt.title("Top 10 Portfolio Holdings")
plt.xlabel("Weight (%)")
plt.tight_layout()
plt.savefig("../reports/top_holdings.png", dpi=300)
plt.show()

# -------------------------------
# Chart 12 - Payment Mode
# -------------------------------
plt.figure(figsize=(7,5))
sns.countplot(data=txn, x="payment_mode")
plt.xticks(rotation=30)
plt.title("Payment Modes")
plt.tight_layout()
plt.savefig("../reports/payment_mode.png", dpi=300)
plt.show()

# -------------------------------
# Chart 13 - KYC Status
# -------------------------------
plt.figure(figsize=(6,5))
sns.countplot(data=txn, x="kyc_status")
plt.title("KYC Status")
plt.tight_layout()
plt.savefig("../reports/kyc_status.png", dpi=300)
plt.show()

# -------------------------------
# Chart 14 - Income Distribution
# -------------------------------
plt.figure(figsize=(8,5))
sns.histplot(txn["annual_income_lakh"], bins=20)
plt.title("Annual Income Distribution")
plt.tight_layout()
plt.savefig("../reports/income_distribution.png", dpi=300)
plt.show()

# -------------------------------
# Chart 15 - Correlation Matrix
# -------------------------------
plt.figure(figsize=(10,8))
corr = txn.select_dtypes(include="number").corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("../reports/correlation_matrix.png", dpi=300)
plt.show()

print("="*50)
print("EDA COMPLETED SUCCESSFULLY")
print("All charts saved inside reports folder.")
print("="*50)
```
